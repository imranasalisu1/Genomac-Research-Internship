```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/cleaned_subsystem_strain_results.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data
print(data.head())
# Display the first few rows of the data
print(data.head())
```

```python
# Prepare the data for the heatmap
# Set the first column as the index (subsystems)
heatmap_data = data.set_index("Subsystem and corresponding genes")

# Convert the data to a numeric matrix
heatmap_data = heatmap_data.astype(float)

# Create the heatmap
plt.figure(figsize=(15, 10))  # Adjust the figure size as needed
sns.heatmap(
    heatmap_data,
    cmap="coolwarm",  # Color gradient (blue to red)
    center=0,  # Center the color map at zero
    robust=True,  # Use robust scaling to handle outliers
    annot=False,  # Set to True if you want to display values in each cell
    fmt=".1f",  # Format for annotations (if annot=True)
    linewidths=0.5,  # Add lines between cells
    cbar_kws={"shrink": 0.8},  # Adjust the color bar size
)
# Add title and labels
plt.title("Heatmap of Subsystem Values Across Strains", fontsize=16, pad=20)
plt.xlabel("Strains", fontsize=12)
plt.ylabel("Subsystems", fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=8)

# Save the heatmap as a PNG file
plt.savefig("subsystem_strain_heatmap.png", dpi=300, bbox_inches="tight")

# Show the plot (optional)
plt.show()

# Message to confirm the file has been saved
print("Heatmap saved as 'subsystem_strain_heatmap.png' in your working directory.")
```
